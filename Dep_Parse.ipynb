```python
import spacy
from spacy import displacy
from spacy.matcher import DependencyMatcher


string = "a tree is hit in the branch with a crushing attack"
nlp = spacy.load("en_core_web_sm")
doc = nlp(string)

matcher = DependencyMatcher(nlp.vocab)
# Only run nlp.make_doc to speed things up
patterns = [{"RIGHT_ID": "anchor_verb",
             "RIGHT_ATTRS": {"DEP": "ROOT"}},

            {"LEFT_ID": "anchor_verb",
             "REL_OP": ">",
            "RIGHT_ID": "subj_verb",
             "RIGHT_ATTRS": {"DEP": "nsubjpass"}},

            {"LEFT_ID": "anchor_verb",
             "REL_OP": ">>",
             "RIGHT_ID": "location_verb",
             "RIGHT_ATTRS": {"DEP": "prep"}}]

matcher.add("Inputs1", [patterns])
matches = matcher(doc)

match_id, token_ids = matches[0]
matched_words = []
for i in range(len(token_ids)):
    print(patterns[i]["RIGHT_ID"] + ":", doc[token_ids[i]].text)
    matched_words.append(doc[token_ids[i]].text)


```
